```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1 How many plots?
def return_sinusoid(x, A, w):
    y =  A * np.sin(w*x)
    return y

x = np.linspace(0, 2 * np.pi, 1000)
A_values = np.array([1.2, 1.4, 1.6, 1.8, 2.0])
w_values = np.array([1, 2, 3, 4, 5])
print ("A_values:", A_values)

plt.figure()
for A in A_values:
   for w in w_values:
        y = return_sinusoid(x, A, w)
        plt.plot(x, y, label = f'A={A}, w={w}')
plt.title('Sinusoidal Waves with Different Amplitudes and Wave Numbers') 
plt.xlabel('x') 
plt.ylabel('y = A * sin(x * w)') 
plt.legend() 
plt.show()
```

```python
#2 Randomness

list_1 = np.random.randint(0, 101, size = 40)
list_2 = np.random.randint(0, 101, size = 40)

x = np.arange(40)
plt.figure(figsize = (10,10))
plt.plot(x, list_1, color = 'orange', linewidth = 10, label = 'List 1 (Orange)')
plt.plot(x, list_2, color = 'red', linestyle = '--', label = 'List 2 (Red)')
plt.title('Randomness')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
